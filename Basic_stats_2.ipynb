```python
import pandas as pd
import numpy as np
```

```python
pat = pd.read_csv('Patient_Data.csv')
pat.info()
```

```python
print(pat.columns)
```

```python
pat_df = pat
pat_df.drop(columns = ['Name','ReceptionistID','CheckInTime'], inplace = True)
pat_df
pat_df.shape
```

```python
pat_df.describe()
```

```python
pat_df.groupby('Department')['BillAmount'].sum()
```

```python
pat_df.drop_duplicates(subset = ['PatientID'], inplace = True)
```

```python
pat_df.shape
```

```python
pat_df.isna()
```

```python
pat_df['BillAmount'].isna().sum()
```

```python
mean = pat_df['BillAmount'].mean()
pat_df['BillAmount'] = pat_df['BillAmount'].fillna(mean)
```

```python
pat_df['BillAmount'].isna().sum()
```

```python
# New patients for the current week
new_patients = pd.DataFrame({
    'PatientID': [106, 107],
    'Department': ['Cardiology', 'Orthopedics'],
    'BillAmount': [5800, 6400],
    'DoctorID': [205, 206]
})
    
# Concatenate row-wise
con_df = pd.concat([pat_df, new_patients], ignore_index=True)
print(con_df)
```
