```python
numeric_cols = sales_df.select_dtypes(include=['int64', 'float64']).columns
numeric_cols
```

```python
categorical_cols = sales_df.select_dtypes(include = ['object']).columns
categorical_cols
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=sales_df[numeric_cols], palette="Set2")
plt.title("Boxplots of Numerical Variables")
plt.xticks(rotation=45)
plt.show()
```

```python
for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(data=sales_df, x=col, palette='Set2', hue=col)
    plt.title(f'Category Distribution: {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
for col in numeric_cols:
    Q1 = sales_df[col].quantile(0.25)
    Q3 = sales_df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = sales_df[(sales_df[col] < lower_bound) | (sales_df[col] > upper_bound)][col]

    print(f"\nColumn: {col}")
    print(f"IQR: {IQR}")
    print(f"Lower Bound: {lower_bound}")
    print(f"Upper Bound: {upper_bound}")
    print(f"Outliers:\n{outliers}")
```

```python
for col in categorical_cols:
    plt.figure(figsize=(8,4))
    sns.countplot(data=sales_df, x=col, palette='Set2',hue = col)
    plt.title(f'Category Distribution: {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
