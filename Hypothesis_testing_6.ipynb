```python
#Objective:
#To investigate the restaurant owners' claim about the increase in weekly operating costs using hypothesis testing.
```

```python
#Data Provided:
#	The theoretical weekly operating cost model: W = $1,000 + $5X
#	Sample of 25 restaurants with a mean weekly cost of Rs. 3,050
#	Number of units produced in a week (X) follows a normal distribution with a mean (μ) of 600 units and a standard deviation (σ) of 25 units
```

```python
import math
from scipy.stats import norm
```

```python
#GIVEN INFORMATION
```

```python
#mu : W=1000+5X #mu->population mean
#n : 25         #n-> sample size
#X_bar : 3,050  #X_bar-> sample size
#μ : 600        #μ-> population mena 
#σ  : 25        #sigma-> standard deviation
#aplha:0.05     #alpha-> level of significant     
```

```python
                 #State the Hypotheses statement:
```

```python
# NULL HYPOTHESIS:
       # mu = 1000+5(600)=4000

#Alternative Hypothesis:
       # mu > 4000
```

```python
                 #Calculate the Test Statistic:
```

```python
#step1
#THEORETICAL MEAN WEEKLY COST mu=1000 +5*600=4000
#mu=4000
```

```python
#step2
# standard deviation of weekly cost 5*25
#sigma=125
```

```python
#step3
#Z-test formula
#Z= X_bar - mu/(sigma/squareroot(n))
#Z= 3050-4000/(125/squareroot(25))
#Z= -950/25
#Z= -38
```

```python
#step4
#critical value(right tailed,alpha=0.05)
#z(critical)=1.65
#Z= -38
```

```python
#step5
#compare z value and critical point
#accept H0
#Z< 1.65
#-38<1.65
#reject H0
#Z> 1.65
```

```python
                                      #CODE
```

```python
import math
from scipy.stats import norm

# Given values
sample_mean = 3050
X_mean = 600
sigma_X = 25
n = 25
alpha = 0.05

# population mean cost
mu = 1000 + 5 * X_mean

# Standard deviation of weekly cost
sigma = 5 * sigma_X

# Z-test calculation
z = (sample_mean - mu) / (sigma / math.sqrt(n))

# Critical value (right-tailed test)
z_critical = norm.ppf(1 - alpha)

print("Hypothesis Testing Results : ")
print("Theoretical Mean Cost (μ):", mu)
print("Sample Mean Cost (x̄) : " ,sample_mean)
print("Z-statistic : " ,z)
print("Critical Z-value:",z_critical)

if z > z_critical:
    print(" Reject the Null Hypothesis")
    print(" Weekly operating costs are higher than the model predicts.")
else:
    print(" Fail to Reject the Null Hypothesis")
    print(" No evidence that weekly operating costs are higher than the model.")
```
