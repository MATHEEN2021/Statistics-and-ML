```python
import scipy.stats as st 
import numpy as np 
from scipy.stats import t, norm 
```

```python
data = np.array([1.13,1.55,1.43,0.92,1.25,1.36,1.32,0.85,
                 1.07,1.48,1.20,1.33,1.18,1.22,1.29])

n = len(data)
mean = np.mean(data)
s = np.std(data, ddof=1)

# (a) t-interval
t_critical = t.ppf(0.995, n-1)
t_result = (mean - t_critical*(s/np.sqrt(n)),
        mean + t_critical*(s/np.sqrt(n)))

# (b) z-interval
sigma = 0.2
z_critical = norm.ppf(0.995)
z_result = (mean - z_critical*(sigma/np.sqrt(n)),
        mean + z_critical*(sigma/np.sqrt(n)))

print("99% CI using sample SD (t):", t_result)
print("99% CI using known SD (z):", z_result)
```

```python
#Result
#99% confident that the true mean durability of the print-heads lies between 1.078 and 1.400 million characters.
```

```python
#Result:
#With 99% confidence, the true mean durability lies between 1.105 and 1.373 million characters when the population standard deviation is known.
```
